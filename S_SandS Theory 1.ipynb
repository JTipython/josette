```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

# Q1: 將兩個旋波相加後，並畫在一起。

```python
phs1 = np.linspace(0, 2*np.pi, 300)

y1 = np.sin(5*phs1)
y2 = 0.7 * np.sin(5*phs1 + 1*np.pi)

plt.plot(y1) # blue
plt.plot(y2) # green
plt.plot(y1 + y2) # red
```

# Q2: 嘗試取樣的不同，會出現怎樣的變化。

```python
t1 = np.linspace(0, 2*np.pi, 50)

y1 = np.sin(5*t1)

plt.plot(y1)
```

# Q3: 試著用不同的量化數模擬出下面的結果。

```python
t = np.linspace(0, 2*np.pi, 300)
y = np.sin(t)
y2 = (y*2).astype(int)
y4 = (y*4).astype(int)

plt.plot(y)
plt.plot(y2)
plt.plot(y4/3.0)
plt.legend(['sine', '2-bit', '3-bit'])
```
